```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("houses_area_price.csv")
```

```python
df
```

```python
#categorical variables: nominal eg(male, female), (green, blue, red), (atlanta, dallas)
#ordinal---> eg(graduate, masters phd), (high, medium, low), kinda have a rank
```

```python
dummies = pd.get_dummies(df.town)
```

```python
dummies
```

```python
merged = pd.concat([df, dummies], axis="columns")#lets merge the two dataframes
```

```python
merged
```

```python
merged.drop("town", axis="columns")
```

```python
final = merged.drop(["town","west windsor"], axis="columns")
```

```python
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
X = final.drop("price", axis="columns")
```

```python
y=final["price"]# y = final.price ---> will also work
```

```python
X
```

```python
y
```

```python
model.fit(X, y)
```

```python
model.predict([[2800, 0, 1]])
```

```python
model.predict([[3400, 0, 0]])
```

```python
model.score(X, y)
```
