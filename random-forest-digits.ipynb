```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_digits
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix

digits = load_digits()
%matplotlib inline
```

```python
plt.gray()
```

```python
df = pd.DataFrame(digits.data)
```

```python
df
```

```python
pd.set_option("display.max_columns", 64)#sets the max number of columns to display
pd.set_option("display.max_rows", 1796)#sets the max number of rows to display
```

```python
df
```

```python
digits.target #displays the target variables
```

```python
df["target"] = digits.target
```

```python
df
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(["target"], axis = "columns"), digits.target, test_size = 0.2 )
```

```python
 len(X_train)
```

```python
len(X_test)
```

```python
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
y_predicted = model.predict(X_test)
```

```python
conf_matr = confusion_matrix(y_test, y_predicted)
```

```python
conf_matr
```

```python
plt.figure(figsize=(10, 7))
sns.heatmap(conf_matr, annot=True)
plt.xlabel = ("Predicted")
plt.ylabel("Truth")
```
