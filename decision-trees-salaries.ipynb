```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
```

```python
df = pd.read_csv("salaries.csv")
```

```python
df.info()
```

```python
inputs = df.drop("Salary> 100k", axis="columns")
target = df["Salary> 100k"]
```

```python
inputs
```

```python
target
```

```python
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
#encodes all the 3 columns
```

```python
inputs["company_n"] = le_company.fit_transform(inputs["company"])
inputs["job_n"] = le_company.fit_transform(inputs["job"])
inputs["degree_n"] = le_company.fit_transform(inputs["degree"])
#the above lines of code adds those encoded columns to the dataset
```

```python
inputs
```

```python
inputs
```

```python
inputs_n = inputs.drop(["company", "job", "degree"], axis = "columns")
```

```python
inputs_n
```

```python
model = tree.DecisionTreeClassifier()
model.fit(inputs_n, target)#train the model
```

```python
model.score(inputs_n, target)#get the score of the model
```

```python
#the score is 1 because we use the same datasets thats used for training and also simple
```
